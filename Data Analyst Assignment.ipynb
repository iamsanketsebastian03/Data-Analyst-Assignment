Lets plot the correlation matrix to get an understanding of how our variables are related to each another.

```python
corr_metrics = batterydata.corr()
corr_metrics.style.background_gradient()
```

A common task is to check if some variables are correlated. You can easily calculate the correlation between each pair of variable, and plot this as a heatmap. This lets you discover which variable are related one each other.

Here we can say that Timestamp and Equivalent cycle ,Grid cumulative energy consumption and Timestamp ,Grid cumulative energy consumption and Equivalent cycle is strongly correlated and so is Grid status and Current.

As per the goal of this exercise we are to find relation between two columns.
We can see that Grid status and SOC have a very low correlation.
Equivalent cycle and SOH are highly correlated but as its a negative number we can say that they are inversely correlated.
SOC and Temperature share a negative low correlation.

```python
sns.relplot(x="Equivalent cycle", y="SOH", kind="line", data=batterydata)
```

```python
sns.regplot(x=batterydata["Equivalent cycle"], y=batterydata["SOH"])
plt.show()
```

As seen from correlation matrix we can confirm from Line plot and scatterplot with Regression line that Equivalent cycle and SOH are correlated.As value of Equivalent cycle increases the value of SOH decreases hence negative correlation.

Lets look at other pairs and see if they are correlated to each other.

```python
sns.scatterplot(x='Grid status', y='SOC', data=batterydata)
```

```python
sns.regplot(x=batterydata["Grid status"], y=batterydata["SOC"])
plt.show()
```

As seen we can conclude that Grid Status and SOC varriable has very low correlation.The regression line in the regplot shows us that there is some positive correlation as seen in the correlation matrix.

```python
sns.regplot(x=batterydata["SOC"], y=batterydata["Temperature"])
plt.show()
```

We can see in this scatterplot that there is low correlation between SOC and Temperature column. While the regression line shows that both the columns are negatively correlated as was seen in correlation matrix.
